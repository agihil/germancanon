```python
# Pakete importieren
import pandas as pd
import plotly.express as px
from scipy.stats import kendalltau
```

# Metzler Literaturgeschichte

```python
#metzler = pd.read_csv("/content/drive/MyDrive/Kanonizität/resources/literaturgeschichten/prepared/litges001.csv", index_col = [0])
```

```python
metzler = pd.read_csv("litges001.csv", sep=";", index_col=0)
metzler_autoren = metzler[metzler["Eintragsart"]=="Autor"].reset_index(drop=True)
metzler_autoren.head()
```

```python
metzler_autoren_short = metzler_autoren.sort_values(by="Anteil_Seiten", ascending=False)[0:100]

fig = px.bar(metzler_autoren_short,
             x="Eintrag",
             y="Anteil_Seiten",
             color_discrete_sequence=["darkblue"],
             title="Seitenanteile der Autor:innen für Metzler Literaturgeschichte (Top 100)")

fig.update_layout(xaxis={'categoryorder':'total descending', \
                  "tickfont" : dict(size=5)})
fig.show()
```

# Brenner

```python
brenner = pd.read_csv("litges003.csv", sep=";", index_col=0)
brenner_autoren = brenner[brenner["Eintragsart"]=="Autor"].reset_index(drop=True)
brenner_autoren.head()
```

```python
brenner_autoren_short = brenner_autoren.sort_values(by="Anteil_Seiten", ascending=False)[0:100]

fig = px.bar(brenner_autoren_short,
             x="Eintrag",
             y="Anteil_Seiten",
             color_discrete_sequence=["darkblue"],
             title="Seitenanteile der Autor:innen für DeGruyter (Top 100)")

fig.update_layout(xaxis={'categoryorder':'total descending', \
                  "tickfont" : dict(size=5)})
fig.show()
```

# Rank Correlation für Top 100

```python
correlation, p_value = kendalltau(brenner_autoren_short['Anteil_Seiten'], metzler_autoren_short['Anteil_Seiten'])
print(f"Kendall correlation: {correlation}, p-value: {p_value}")
```

```python
metzler_autoren["Anteil_Seiten_Metzler"] = metzler_autoren["Anteil_Seiten"]
brenner_autoren["Anteil_Seiten_Brenner"] = brenner_autoren["Anteil_Seiten"]

brenner_small = brenner_autoren.sort_values(by="Anteil_Seiten_Brenner", ascending=False)[0:100].reset_index(drop=True)
metzler_small = metzler_autoren.sort_values(by="Anteil_Seiten_Metzler", ascending=False)[0:100].reset_index(drop=True)

tops = brenner_small.merge(metzler_small, on="Eintrag", how="inner")


tops.head()
```

```python
brenner_small
```

```python
# Rang-Korrelationen für verschiedene Tops
metzler_autoren["Anteil_Seiten_Metzler"] = metzler_autoren["Anteil_Seiten"]
brenner_autoren["Anteil_Seiten_Brenner"] = brenner_autoren["Anteil_Seiten"]

correlations = {i:0 for i in range(1,500)}

for i in correlations.keys():
  brenner_small = brenner_autoren.sort_values(by="Anteil_Seiten_Metzler", ascending=False)[0:i].reset_index(drop=True)
  metzler_small = metzler_autoren.sort_values(by="Anteil_Seiten_Brenner", ascending=False)[0:i].reset_index(drop=True)
  tops = pd.concat([brenner_small, metzler_small])
  tops = tops.drop_duplicates()

  correlation, p_value = kendalltau(brenner_small['Anteil_Seiten'], metzler_small['Anteil_Seiten'])

  correlations[i] = correlation
```

```python
correlation, p_value = kendalltau([0,1], [1,0])
correlation, p_value = kendalltau([1,0], [1,0])
correlation
```

```python
kendall_correlations = pd.DataFrame({"Top": correlations.keys(), "Correlation": correlations.values()})
kendall_correlations.head()
```

```python
fig = px.line(kendall_correlations,
              x="Top",
              y="Correlation",
              color_discrete_sequence=["darkblue"],
              title="Kendall Korrelationskoeffizient zwischen Metzler und DeGruyter Literaturgeschichte")
fig.show()
```
